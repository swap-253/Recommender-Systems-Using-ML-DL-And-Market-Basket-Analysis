```python
#### preprocessing of title and plot columns
import regex as re
movies_data33['cleaned_title'] = movies_data33['title_without_year'].apply(lambda x: x.lower())
movies_data33['cleaned_title'] = movies_data33['cleaned_title'].apply(lambda x: re.sub('[^A-Za-z0-9]+', ' ', x))
movies_data33['cleaned_overview'] = movies_data33['plot'].apply(lambda x: x.lower())
movies_data33['cleaned_overview'] = movies_data33['cleaned_overview'].apply(lambda x: re.sub('[^A-Za-z0-9]+', ' ', x))
movies_data33.head()
```

```python
### utilizing bert enodings for title column(128 D)
import tensorflow_hub as hub
import tensorflow as tf
import tensorflow_text as text
preprocessor = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-2_H-128_A-2/1", trainable=True)
import tensorflow as tf

def get_bert_embeddings(text, preprocessor, encoder):

  text_input = tf.keras.layers.Input(shape=(), dtype=tf.string)
  encoder_inputs = preprocessor(text_input)
  outputs = encoder(encoder_inputs)
  embedding_model = tf.keras.Model(text_input, outputs['pooled_output'])
  sentences = tf.constant([text])
  return embedding_model(sentences)

movies_data33['encodings_title'] = movies_data33['cleaned_title'].apply(lambda x: get_bert_embeddings(x, preprocessor, encoder))
```

```python
### utilzing wiki trained bert encoding for plot column(768 D)
encoder2 = hub.KerasLayer("https://tfhub.dev/google/experts/bert/wiki_books/2", trainable=True)
movies_data33['encodings_overview'] = movies_data33['cleaned_overview'].apply(lambda x: get_bert_embeddings(x, preprocessor, encoder2))
```

### In the below function, we ask the user to input user id and basis the plot, we recommend 15 most similar movies that user hasn't watched where bert plot embedding capture atleast a 60 percent similarity and atleast 270  users have watched the movies with it having an avg rating of over and above 3.5

### In the below function, we ask the user to input user id and basis the plot, for each of the movies we recommend 15 most similar movies that user hasn't watched where bert plot embedding capture atleast a 60 percent similarity and atleast 270  users have watched the movies with it having an avg rating of over and above 3.5. Now we sort those movies by how many times have they been captured and by their average ratings. Basis that we offer a ranked recommendation on what user might like


```python
subset_rec_movies=movies_ratings_v1[(movies_ratings_v1['cleaned_plot']!='none')&
  (movies_ratings_v1['avg_rating']>=3.5)&(movies_ratings_v1['num_users']>=270)]['sorted_movie_id'].unique()
def recommend_for_user_id_ranked():
  dict_movie_counts,dict_movie_ratings={},{}
  text = int(input())
  high_rated_user_movies=ratings_data_v2[(ratings_data_v2['UserID']==text)&(ratings_data_v2['cleaned_plot']!='none')&(ratings_data_v2['Rating']>=4)]['sorted_movie_id'].unique()
  if len(high_rated_user_movies)==0:
    return "UserID doesn't exist"
  high_rated_user_movies_df=ratings_data_v2[(ratings_data_v2['UserID']==text)&(ratings_data_v2['cleaned_plot']!='none')&(ratings_data_v2['Rating']>=4)][['UserID','MovieID','sorted_movie_id','Title','Rating','final_genre','cast','director','cleaned_plot']]
  movies_watched_by_user=ratings_data_v2[(ratings_data_v2['UserID']==text)]['sorted_movie_id'].unique()
  for i in high_rated_user_movies:
    movies_to_recommend=movie_sim_df4[(movie_sim_df4['movie1']==i)&(movie_sim_df4['sent_bert_plot_sim']>=0.6)&
    ~(movie_sim_df4['movie2'].isin(movies_watched_by_user))&(movie_sim_df4['movie2'].isin(subset_rec_movies))].sort_values(by=['sent_bert_plot_sim'],
    ascending=False)[:15][['movie2','sent_bert_plot_sim']].drop_duplicates().reset_index()
    for j in range(len(movies_to_recommend)):
      if movies_to_recommend['movie2'][j] not in dict_movie_counts.keys():
        dict_movie_counts[movies_to_recommend['movie2'][j]]=1
        dict_movie_ratings[movies_to_recommend['movie2'][j]]=movies_to_recommend['sent_bert_plot_sim'][j]
      else:
        dict_movie_counts[movies_to_recommend['movie2'][j]]+=1
        dict_movie_ratings[movies_to_recommend['movie2'][j]]+=movies_to_recommend['sent_bert_plot_sim'][j]
  df_counts=pd.DataFrame()
  df_sim=pd.DataFrame()
  df_counts['movie2']=dict_movie_counts.keys()
  df_counts['freq']= dict_movie_counts.values()
  df_sim['movie2']=dict_movie_ratings.keys()
  df_sim['similarity_sum']=dict_movie_ratings.values()
  df_merged=df_counts.merge(df_sim,on=['movie2'],how='left')
  df_merged['avg_similarity']=df_merged['similarity_sum']/df_merged['freq']
  # movies_to_recommend_sim=movie_sim_df4[(movie_sim_df4['movie1'].isin(high_rated_user_movies))&(movie_sim_df4['sent_bert_plot_sim']>=0.6)&
  # ~(movie_sim_df4['movie2'].isin(movies_watched_by_user))&(movie_sim_df4['movie2'].isin(subset_rec_movies))].sort_values(by=['sent_bert_plot_sim'],ascending=False)[:15]
  sorted_movies_by_sim_counts_and_ratings=df_merged.sort_values(by=['freq','avg_similarity'],ascending=False)[:15]
  sorted_movies_by_sim_counts_and_ratings['rank']=[i for i in range(1,len(sorted_movies_by_sim_counts_and_ratings)+1)]
  # sorted_movies_by_sim_counts_and_ratings_movies=sorted_movies_by_sim_counts_and_ratings['movie2'].unique()
  ratings_data_v2_copy=ratings_data_v2.copy()
  ratings_data_v3_copy=ratings_data_v2_copy.merge(sorted_movies_by_sim_counts_and_ratings,left_on=['sorted_movie_id'],right_on=['movie2'],how='left')
  rec_movies_df=ratings_data_v3_copy[ratings_data_v3_copy['movie2'].notnull()].sort_values(by=['rank'])[['sorted_movie_id','Title','cast','director','final_genre','cleaned_plot']].drop_duplicates(keep='first')
  return high_rated_user_movies_df,sorted_movies_by_sim_counts_and_ratings.drop(['similarity_sum'],axis=1),rec_movies_df
```

```python
subset_rec_movies=movies_ratings_v2[(movies_ratings_v2['avg_rating']>=3.5)&(movies_ratings_v2['num_users']>=270)]['mov_ids_index'].unique()
def recommend_for_user_id():
  text = int(input())
  high_rated_user_movies=ratings_data_v3[(ratings_data_v3['UserID']==text)].sort_values(by=['Rating'],ascending=False)[:15]['mov_ids_index'].unique()
  if len(high_rated_user_movies)==0:
    return "UserID doesn't exist"
  high_rated_user_movies_df=ratings_data_v3[(ratings_data_v3['UserID']==text)].sort_values(by=['Rating'],
  ascending=False)[:15][['UserID','MovieID','mov_ids_index','Title','Rating','final_genre','cast','director']]
  movies_watched_by_user=ratings_data_v3[(ratings_data_v3['UserID']==text)]['mov_ids_index'].unique()
  movies_to_recommend=movies_sim_df[(movies_sim_df['movie1'].isin(high_rated_user_movies))&(movies_sim_df['sim_score']>=0.3)&
  ~(movies_sim_df['movie2'].isin(movies_watched_by_user))&(movies_sim_df['movie2'].isin(subset_rec_movies))].sort_values(by=['sim_score'],ascending=False)[:15]['movie2'].unique()
  movies_to_recommend_sim=movies_sim_df[(movies_sim_df['movie1'].isin(high_rated_user_movies))&(movies_sim_df['sim_score']>=0.3)&
  ~(movies_sim_df['movie2'].isin(movies_watched_by_user))&(movies_sim_df['movie2'].isin(subset_rec_movies))].sort_values(by=['sim_score'],ascending=False)[:15]
  rec_movies_df=ratings_data_v3[ratings_data_v3['mov_ids_index'].isin(movies_to_recommend)][['mov_ids_index','Title','cast','director','final_genre']].drop_duplicates(keep='first')
  return high_rated_user_movies_df,movies_to_recommend,movies_to_recommend_sim,rec_movies_df
```

```python
subset_rec_movies=movies_ratings_v2[(movies_ratings_v2['avg_rating']>=3.5)&(movies_ratings_v2['num_users']>=270)]['mov_ids_index'].unique()
def recommend_for_user_id_ranked_tfidf():
  dict_movie_counts,dict_movie_ratings={},{}
  text = int(input())
  high_rated_user_movies=ratings_data_v3[(ratings_data_v3['UserID']==text)&(ratings_data_v3['Rating']>=4)]['mov_ids_index'].unique()
  if len(high_rated_user_movies)==0:
    return "UserID doesn't exist"
  high_rated_user_movies_df=ratings_data_v3[(ratings_data_v3['UserID']==text)&(ratings_data_v3['Rating']>=4)][['UserID','MovieID','mov_ids_index','Title','Rating','final_genre','cast','director']]
  movies_watched_by_user=ratings_data_v3[(ratings_data_v3['UserID']==text)]['mov_ids_index'].unique()
  for i in high_rated_user_movies:
    movies_to_recommend=movies_sim_df[(movies_sim_df['movie1']==i)&(movies_sim_df['sim_score']>=0.3)&
    ~(movies_sim_df['movie2'].isin(movies_watched_by_user))&(movies_sim_df['movie2'].isin(subset_rec_movies))].sort_values(by=['sim_score'],
    ascending=False)[:15][['movie2','sim_score']].drop_duplicates().reset_index()
    for j in range(len(movies_to_recommend)):
      if movies_to_recommend['movie2'][j] not in dict_movie_counts.keys():
        dict_movie_counts[movies_to_recommend['movie2'][j]]=1
        dict_movie_ratings[movies_to_recommend['movie2'][j]]=movies_to_recommend['sim_score'][j]
      else:
        dict_movie_counts[movies_to_recommend['movie2'][j]]+=1
        dict_movie_ratings[movies_to_recommend['movie2'][j]]+=movies_to_recommend['sim_score'][j]
  df_counts=pd.DataFrame()
  df_sim=pd.DataFrame()
  df_counts['movie2']=dict_movie_counts.keys()
  df_counts['freq']= dict_movie_counts.values()
  df_sim['movie2']=dict_movie_ratings.keys()
  df_sim['similarity_sum']=dict_movie_ratings.values()
  df_merged=df_counts.merge(df_sim,on=['movie2'],how='left')
  df_merged['avg_similarity']=df_merged['similarity_sum']/df_merged['freq']
  # movies_to_recommend_sim=movie_sim_df4[(movie_sim_df4['movie1'].isin(high_rated_user_movies))&(movie_sim_df4['sent_bert_plot_sim']>=0.6)&
  # ~(movie_sim_df4['movie2'].isin(movies_watched_by_user))&(movie_sim_df4['movie2'].isin(subset_rec_movies))].sort_values(by=['sent_bert_plot_sim'],ascending=False)[:15]
  sorted_movies_by_sim_counts_and_ratings=df_merged.sort_values(by=['freq','avg_similarity'],ascending=False)[:10]
  sorted_movies_by_sim_counts_and_ratings['rank']=[i for i in range(1,len(sorted_movies_by_sim_counts_and_ratings)+1)]
  # sorted_movies_by_sim_counts_and_ratings_movies=sorted_movies_by_sim_counts_and_ratings['movie2'].unique()
  ratings_data_v2_copy=ratings_data_v3.copy()
  ratings_data_v3_copy=ratings_data_v2_copy.merge(sorted_movies_by_sim_counts_and_ratings,left_on=['mov_ids_index'],right_on=['movie2'],how='left')
  rec_movies_df=ratings_data_v3_copy[ratings_data_v3_copy['movie2'].notnull()].sort_values(by=['rank'])[['mov_ids_index','Title','cast','director','final_genre']].drop_duplicates(keep='first')
  return high_rated_user_movies_df,sorted_movies_by_sim_counts_and_ratings.drop(['similarity_sum'],axis=1),rec_movies_df
```
